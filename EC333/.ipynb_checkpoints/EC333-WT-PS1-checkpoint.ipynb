```python
import pandas as pd

# Specify the file path to the .dta file
file_path = '/Users/danielseymour/Downloads/TeachingRatings.dta'

# Load the .dta file into a pandas DataFrame
try:
    teaching_ratings = pd.read_stata(file_path)
    # Display the first few rows of the DataFrame
    print(teaching_ratings.head())
except FileNotFoundError:
    print(f"File not found: {file_path}. Please check the path and try again.")
    
```

```python
teaching_ratings.describe()
```

```python
#Question 2.a
import statsmodels.api as sm

# Define the dependent variable (course_eval) and independent variable (beauty)
X = teaching_ratings['beauty']  # Independent variable
y = teaching_ratings['course_eval']  # Dependent variable

# Add a constant to the independent variable (intercept term)
X = sm.add_constant(X)
print(X)
# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression results
print(model.summary())
```

```python
#Question 2.b
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
teaching_ratings['std_beauty'] = scaler.fit_transform(teaching_ratings[['beauty']])
teaching_ratings['std_course_eval'] = scaler.fit_transform(teaching_ratings[['course_eval']])

X = sm.add_constant(teaching_ratings['std_beauty'])
y = teaching_ratings['std_course_eval']
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
#Question 2.c
# Define the dependent variable (course_eval) and independent variable (beauty)
X = teaching_ratings['beauty']  # Independent variable
y = teaching_ratings['course_eval']  # Dependent variable

# Add a constant to the independent variable (intercept term)
X = sm.add_constant(X)
# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression results
print(model.summary())

# Extract the p-value of the 'beauty' coefficient
p_value = model.pvalues['beauty']

if p_value < 0.05:
    print(f"The coefficient for 'beauty' is significantly different from zero (p = {p_value:.4f}).")
else:
    print(f"The coefficient for 'beauty' is not significantly different from zero (p = {p_value:.4f}).")
```

```python
#Question 2.d

# Define the dependent variable (course_eval) and independent variables (beauty and female)
X = teaching_ratings[['beauty', 'female']]  # Independent variables
y = teaching_ratings['course_eval']  # Dependent variable

# Add a constant to the independent variables (intercept term)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression results
print(model.summary())
```
